# Missing Values Treatment

```python
# this is for 'Age' column  i.e., numerical data column, we replaced with median of the column
median1=df["Age"].median()
df['Age']=df['Age'].fillna(median1)
```

```python
# this is for 'Embarked' column i.e., categorical (string-object) data column, we replaced with the mode of the column
mode1=df["Embarked"].mode()[0]
df['Embarked']=df['Embarked'].fillna(mode1)
```

```python
df.isnull().sum()
```

median aur mode ke jariye hamne missing values waley columns i., 'Age', and 'Embarked' ko fill kar diya hai

# Encoding (replace categorical data i.e., string-object data to numbers i.e., 0 and 1)
this encoding is called 'one - hot encoding', creating dummies of the categorical columns i.e., Sex and Embarked

```python
# iski madad se humne string-object ko numbers mein encode kr diya hai.
df=pd.get_dummies(df,columns=['Sex','Embarked'])
df.dtypes
```

```python
df.isnull().sum()
```

Note :- now there is no missing value in our dataset and we removed unwanted columns from the dataset.

```python
# split the dataset into dependent variable and independent variable
x=df.drop(['Survived'],axis=1) # all independet variables except 'Survived' placed on x-axis
y=df[['Survived']] # dependent variable placed on y-axis
```

```python
# random_state=42 is default value for all the models.
# now split data into different variables.
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# creating/initiating model
model_lrc=LogisticRegression()
```

```python
# fitting/training model
model_lrc.fit(x_train,y_train)
```

```python
# checking train model score
model_lrc.score(x_train,y_train)
```

```python
# checking test model score
model_lrc.score(x_test,y_test)
```

```python
# random_state ki value = 0 hai tab training mein 82 % aa rha hai lekin tesing mein 75 % score joki 5% accuracy score se jada i.e., 7% hai.
# isliye hamara model bad predictions karega.
# agar hum random_state ki value change karein to score theek ho jaye.
# kisi bhi ml model ki random_state ki default value = 42 hoti hai.
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# dekho random state ki value = 42 karte hi score mein difference aa gaya.
# ab hamare model ki accuracy score training mein 82 % aur tesing mein 77 % hogayi hai joki 5% accuracy score ke barabar hai i.e., 5% hai.
# isliye ab hamara model good predictions karega.
```
