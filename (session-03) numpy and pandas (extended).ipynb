# **Pandas**

In the previous session we hav already defined dataset, we are just importiong that dataset and working on it.
But now here we are creating datasets.
when we are supposed to create only one column in any dataset then we use "pd.Series" and if we are supposed to create multiple columns in any dataset then we use "pd.DataFrames" .

```python
import pandas as pd
```

```python
games1=pd.Series([101,202,303,404,505],index=['cricket','basketball','baseball','football','golf'])
games1
# this method generates a single column dataset in which numeric value is the value which stored in series's single column and
# index value is serial number types value.
# if we don't specify index value then series method will automatically takes index of (0,1,2,3,4,5,6.....) as serial number.
```

```python
games1['cricket']
```

```python
games2=pd.Series([10,20,30,40,50,60],index=['chess','carrom','baseball','golf','cricket','freefire'])
games2
```

```python
games1+games2
```

here NaN values are showing because jo bhi index value (serial number) dono series mein same hain unki values add ho gayi aur jo nahi hain unki values add na hoke NaN show ho gaya hai

```python
df1=pd.DataFrame(np.random.rand(6,3), index='a b c d e f'.split(), columns='score1 score2 score3'.split())
df1
# here we are creating random values for our dataset using 'np.random.rand(no. of rows, no. of columns)' method.
# creating searial number using index.
# creating column names for our dataset using columns variable
# most important we use .split() method to split all the rows and columns names because we didn't declare each row and column
# as separate array that's why.
# index and column values are depend on the argument list of random.rand() method.
# ---------------------------------------------------------------------------------------------------------------------------- #
# no. of rows = index values
# no. of columns = columns values
```

```python
df1["score1"]
```

```python
df1[['score1', 'score2']]
```

```python
df1['score4']=df1['score1']*df1['score2']
df1
```

```python
df['name']
```

```python
df[['name','address']]
# jab ek se jada columns ko print krna ho ya fetch krna ho to double large brackets ka use krte hain.
```
