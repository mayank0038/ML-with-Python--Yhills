# Outliner Treatment

### Below code is for 'Age' column

```python
sns.boxplot(y='Age',data=df)
```

```python
q1=df['Age'].quantile(0.25) # 0.25 value 'age' column ke statistical values waley code se aayi hai.
q3=df['Age'].quantile(0.75) # 0.75 value 'age' column ke statistical values waley code se aayi hai.
iqr=q3-q1
```

```python
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
```

```python
lower_limit
```

```python
upper_limit
```

```python
df=df[(df['Age']>=lower_limit)&(df["Age"]<=upper_limit)]
# "sns.boxplot" mein se outliers ko remove krne ke liye hmne ye code likha hai
# this means data is not removed but outliers are pushed to 'lower_limit' and 'upper_limit 'as shown in below box plot.
```

```python
sns.boxplot(df['Age'])
plt.show()
```

below statements are representation of boxplot lines (top to bottom) :-
* upper_ limit
* q3
* median
* q1
* lower_limit

# Missing Values Treatment

```python
median1=df["Age"].median()
df['Age']=df['Age'].fillna(median1)
# this is for 'Age' column i.e., numerical data column
```

```python
mode1=df["Cabin"].mode()[0]
df['Cabin']=df['Cabin'].fillna(mode1)
# this is for 'Cabin' column i.e., categorical (string-object) data column
```

```python
mode1=df["Embarked"].mode()[0]
df['Embarked']=df['Embarked'].fillna(mode1)
# this is for 'Embarked' column i.e., categorical (string-object) data column
```

```python
df.isnull().sum()
```

median aur mode ke jariye hamne missing values waley columns i., 'Age', 'Cabin' and 'Embarked' ko fill kar diya hai

# Encoding (replace categorical data i.e., string-object data to numbers i.e., 0 and 1)
