here horsepower is showing 2 different sides i.e., integer in *df.head()* code and string-object in *df.info()*

```python
df['horsepower'].value_counts()
```

in the above code you will find '?' sign in the output. Bas vahi hain hamarey hidden missing values. jinhe hamein handle krna hai

```python
df['horsepower']=df['horsepower'].replace('?', np.nan)
df['horsepower']=df['horsepower'].astype(float)
median1=df['horsepower'].median()
median1
df['horsepower'].replace(np.nan,median1,inplace=True)
```

uper ke is code se hmne horsepower ke dtype ko *string-object* se *float64* mein change kr diya hai aur sare NaN values ya bolein to hidden missing values ko meadian se replace kar diya hai.
