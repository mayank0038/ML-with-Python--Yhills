```python
# step 1. import library
# -----------------------------------------------------------------------------------------------------------------------------------------------------
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression # jo bhi model ham create karenge uske liye ye code likhna hai
from sklearn.model_selection import train_test_split # aur ye bhi
from sklearn import metrics # model ka score check krne ke liye ye line of code likhenge.
```

```python
# step 2. load dataset
# -----------------------------------------------------------------------------------------------------------------------------------------------------
df=pd.read_csv('/content/drive/MyDrive/Datasets-main/pima-indians-diabetes-2.csv')
df.sample(10)
```

```python
# step 3. check for shape, dtypes and null values
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# step 3. check for shape
df.shape
```

```python
# step 3. check for dtypes
df.dtypes
```

```python
# step 3. check for null/NaN values
df.isnull().sum()
```

```python
# step 4. if dtypes and null/NaN values or hidden missing values (?-symbol) are present,
# then handle using median and mean (for numerical columns) and using mode (for categorical columns).
# step 5. after handelling null/NaN values and '?' symbol values, recheck for dtypes and null/NaN values.
# step 6. one hot encoding, if required --> here no missing value --> no encoding.
# -----------------------------------------------------------------------------------------------------------------------------------------------------
# hence, we skip step 4, step 5 and step 6
# -----------------------------------------------------------------------------------------------------------------------------------------------------
```

```python
# filled with median of every particular column
median1=df['Preg'].median()
df['Preg']=df['Preg'].fillna(median1)
median2=df['Plas'].median()
df['Plas']=df['Plas'].fillna(median1)
median3=df['Pres'].median()
df['Pres']=df['Pres'].fillna(median1)
median4=df['skin'].median()
df['skin']=df['skin'].fillna(median1)
median5=df['test'].median()
df['test']=df['test'].fillna(median1)
median6=df['mass'].median()
df['mass']=df['mass'].fillna(median1)
median7=df['pedi'].median()
df['pedi']=df['pedi'].fillna(median1)
median8=df['age'].median()
df['age']=df['age'].fillna(median1)
median9=df['class'].median()
df['class']=df['class'].fillna(median1)
```

```python
# rechecking of dtypes
df.dtypes
```

```python
# rechecking of null/NaN values
df.isnull().sum()
```

# iske baad humein data-visualization ke concepts ko bhi krna padega.

uske baad hi data splitting ka kaam karenge

```python
# split the dataset into dependent variable and independent variable
x=df.drop(['class'],axis=1) # all independet variables except 'Survived' placed on x-axis
y=df[['class']] # dependent variable placed on y-axis
```

```python
# random_state=42 is default value for all the models.
# now split data into different variables.
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# creating/initiating model
model_lrc=LogisticRegression()
```

```python
# fitting/training model
model_lrc.fit(x_train,y_train)
```

```python
# checking train model score
model_lrc.score(x_train,y_train)
```

```python
# checking test model score
model_lrc.score(x_test,y_test)
```

### difference of model_score is less than 5% (which is general) i.e., 4% (0.78 = 78% for train model and 0.74 = 74% for test model). Hence, our is going for good predictions.
