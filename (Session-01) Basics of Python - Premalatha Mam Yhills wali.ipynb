#sets

```python
sample_set ={1,3,4,35,11,20,2,3,1,"mayank",'manendra',}
```

```python
sample_set
```

hence
1. sets do not allow duplicate values
2. also output will be in some specific order i.e., order for numbers will according to first digit only and order for strings will according to alphabetical only.
3. sets are enclosed within curly braces.

```python
sample_set[0]
```

why error? because sets don't allow to fetch data using index number

```python
sample_set[0]=2.001
```

similarly, we can't change any value in the sets using index. In this case, Sets are immutable

```python
sample_set.add('varshney')
sample_set
```

But we can add any value of any datatype in the Sets. These added values will take place according to their order as we discussed above.

```python
sample_set.remove(11)
sample_set
```

Similarly, we can remove any value of any datatype in the Sets.

hence, sets are mutable according to this concepts as mentioned here and above cell
