```python
import pandas as pd
# these two lines that are mentioned below are compulsory to write so that all the rows and columns of the dataset we could see.
# if we don't write below lines then after some of the columns dot dot dot are shown to us.
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# reading data from dataset in our local machine
# we can read data by uploading our dataset file in jupyter notebook.
# if we not write 'r' at the beginning of the dataset file path then we have to change backslashes (\) to forward-slashes (/).
# here 'df' is a variabe name for our dataset nad '.read' is a function in 'pandas' library to read data from any dataset file.
# we have given '_csv' beacause our file is in '.csv' format.
df=pd.read_csv('/content/drive/MyDrive/Datasets-main/CardioGoodFitness-1.csv')
```

```python
# '.head()' gives only first 5 rows of our dataset, if we don't specify any parameter.
# if we specify the number of rows as the parameter of the method, it shows those specific number of rows.
# if we do not call '.head()' then it will show all the rows and columns of our dataset.
# '.head()' is basically used to check whether our dataset has colums and rows or not.
df.head(10)
```

```python
# '.tail()' gives last 5 rows of our dataset
# it basically does the opposite of '.head()'.
df.tail(10)
```

```python
# it gives us a random row from our dataset.
# if we specify a number as a parameter, then it will show those specific number of rows that are random from our dataset.
df.sample(5)
```

```python
# it basically shows the total number of (rows, columns) present in our dataset
df.shape
```

```python
# it basically shows the datatypes of all the columns present in our datset.
# here for string type names it gives 'object' datatype and for number it gives 'int64' datatype
df.dtypes
```

```python
# it basically shows all the information about our dataset i.e.,
# there is any blank column present or not etc. types information.
df.info()
```

```python
# check if any null or blank column is present in our dataset or not.
df.isnull().sum()
```

```python
# it basically shows different mathematical values of our dataset.
# it only shows the mathematical values of those rows who has numbers of any type.
df.describe()
```

```python
# if we want all the description of dataset then :-
# here 'NaN' means 'Not a Number'.
# NaN is showing beacuse those values can't be calculated.
df.describe(include='all')
```
