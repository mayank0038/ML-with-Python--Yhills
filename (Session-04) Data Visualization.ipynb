```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
titanic_df=pd.read_csv("/content/drive/MyDrive/Datasets-main/titanic-training-data.csv")
```

```python
titanic_df
```

# **Countplot**

```python
# setting the figure, size
plt.figure(figsize=(8,6))
# The parameters of this method defines the size of plot at x-axis and y-axis respectively in inches.
                                        # OR
# The size in inches of the figure to create
# --------------------------------------------------------------------------------------------------------------------------- #
# creating countplot of passengers by class
sns.countplot(x='SibSp', hue='Sex', data=titanic_df)
# this plot works with any type of data.
# in this method the value of x is the name of column which is shown in countplot.
# in this method agar hum 'x' ki jagah 'y' kar dein to hamara countplot graph y-axis se banna start hoga.
# aur ek baat aur ismein hum 'x' aur 'y' ko ek sath pass nahi kar sakte hain.
# ha lekin hum 'x' ke sath ya 'y' ke sath 'hue' ko define kar sakte hain.
# 'hue' kisi particular column ke values ko multiple color ke boxes mein divide kr deta hai.
# ''x', 'y' aur 'hue' ko hum jadatar ese columns ke liye istemal krte hain jismein do ya tin ya chaar tak multiple values
# hotey hain.
# is samay to x-axis se ban rha hai.
# and the value of data is the name of that variable where we import our dataset in the notebook.
# --------------------------------------------------------------------------------------------------------------------------- #
# setting the title and axis labels for out plot
plt.title('Passengers by class and Age')
plt.xlabel('class and age')
plt.ylabel('Count')
# --------------------------------------------------------------------------------------------------------------------------- #
# showing the plot
plt.show()
```

# **Violin Plot**

```python
# setting the figure size
plt.figure(figsize=(4,3))
# The parameters of this method defines the size of plot at x-axis and y-axis respectively in inches.
                                        # OR
# The size in inches of the figure to create
# --------------------------------------------------------------------------------------------------------------------------- #
# creating violinplot of passengers by class
sns.violinplot(x='Survived', y='Age', data=titanic_df)
# violinplot mein hum vo hi columns ko le skte hain jinmein numeric values stored ho.
# in this method the value of 'x' aur 'y' is the name of column which is shown in violinplot.
# in this method agar hum 'x' ki jagah 'y' aur 'y' ki jagah 'x' kar dein to hamara countplot graph accordingly banna start hoga.
# aur ek baat aur ismein hum 'x' aur 'y' ko ek sath pass kar sakte hain.
# and the value of data is the name of that variable where we import our dataset in the notebook.
# --------------------------------------------------------------------------------------------------------------------------- #
# setting the title and axis labels for out plot
plt.title('Violin Plot of Titanic Dataset')
plt.xlabel('Survived Passenger')
plt.ylabel('Age of the passenger')
# --------------------------------------------------------------------------------------------------------------------------- #
# showing the plot
plt.show()
```

# **Box Plot**

```python
# setting the figure size
plt.figure(figsize=(4,3))
# The parameters of this method defines the size of plot at x-axis and y-axis respectively in inches.
                                        # OR
# The size in inches of the figure to create
# --------------------------------------------------------------------------------------------------------------------------- #
# creating boxplot of passengers by class
sns.boxplot(x='Survived', y='Age', data=titanic_df)
# --------------------------------------------------------------------------------------------------------------------------- #
# setting the title and axis labels for out plot
plt.title('Box Plot of Titanic Dataset')
plt.xlabel('Survived Passenger')
plt.ylabel('Age of the passenger')
# --------------------------------------------------------------------------------------------------------------------------- #
# showing the plot
plt.show()
```

# **Swarmplot**

```python
# setting the figure size
plt.figure(figsize=(4,3))
# The parameters of this method defines the size of plot at x-axis and y-axis respectively in inches.
                                        # OR
# The size in inches of the figure to create
# --------------------------------------------------------------------------------------------------------------------------- #
# creating swarmplot of passengers by class
sns.swarmplot(x='Survived', y='Age', data=titanic_df)
# --------------------------------------------------------------------------------------------------------------------------- #
# setting the title and axis labels for out plot
plt.title('Swarm Plot of Titanic Dataset')
plt.xlabel('Survived Passenger')
plt.ylabel('Age of the passenger')
# --------------------------------------------------------------------------------------------------------------------------- #
# showing the plot
plt.show()
```

# **histplot**

```python
# setting the figure size
plt.figure(figsize=(4,3))
# The parameters of this method defines the size of plot at x-axis and y-axis respectively in inches.
                                        # OR
# The size in inches of the figure to create
# --------------------------------------------------------------------------------------------------------------------------- #
# creating histplot of passengers by class
sns.histplot(x='Age', y='Pclass', data=titanic_df, color='red')
# --------------------------------------------------------------------------------------------------------------------------- #
# setting the title and axis labels for out plot
plt.title('Hist Plot of Titanic Dataset')
plt.xlabel('Age Passenger')
plt.ylabel('Pclass of the passenger')
# --------------------------------------------------------------------------------------------------------------------------- #
# showing the plot
plt.show()
```

```python
titanic_df.hist(figsize=(10,10), color='red')
plt.title('Histplot of titatic_dataframe')
plt.show() # by this shows the plot
```

# **Correlation and heatmap of dataset**

```python
corr=titanic_df.corr()

# setting the figure size
plt.figure(figsize=(4,3))
# The parameters of this method defines the size of plot at x-axis and y-axis respectively in inches.
                                        # OR
# The size in inches of the figure to create
# --------------------------------------------------------------------------------------------------------------------------- #
# creating heatmap of passengers by class
sns.heatmap(corr, annot=True)
# --------------------------------------------------------------------------------------------------------------------------- #
# setting the title and axis labels for our plot
plt.title('Heatmap of Titanic Dataset')
plt.xlabel('Features')
plt.ylabel('Features')
# --------------------------------------------------------------------------------------------------------------------------- #
# showing the plot
plt.show()
```
