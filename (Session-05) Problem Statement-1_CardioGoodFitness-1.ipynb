```python
# now we are finding the exact values of outliers using a formula i.e.,
# all the values between q3+1.5*iqr and q1-1.5*iqr are outlier values.
# in the formula :-
# q1 (first quartile) = value of 25% of particular numeric column in dataset.
# q3 (third quartile) = value of 75% of particular numeric column in dataset.
# iqr (inter quartile range) = q3 - q1

# here we finding outliers for 'Age' column
q1 = 24
q3 = 33
iqr = q3-q1
iqr
# here we find values of q1 and q3 from statistical table we find earlier.
```

```python
q1-1.5*iqr
```

```python
q3+1.5*iqr
```

for age column, this means more than 46.5 all the values are outliers.
similarly, less than 10.5 all the values are outliers.
