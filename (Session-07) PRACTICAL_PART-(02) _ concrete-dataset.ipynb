# To find relationship between strength column with other column/attributes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
# pd.set_option('display.max_rows', None) # this will show all rows of dataset
# pd.set_option('display.max_columns', None) # this will show all columns of dataset
# we will split our data into train data and test data.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets-main/concrete.csv')
df.sample(10)
```

 in this dataset 'strength' is the dependent variable and others are independent variables.

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
sns.pairplot(df,diag_kind='kde')
# to study the co-relation between the attributes, we create a pairplot.
```

```python
# dependent variable -----------------
y=df[['strength']]
# independent variables --------------
X=df.drop(['strength'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
model_1 = LinearRegression()
model_1.fit(X_train, y_train)
```

```python
# here we are checking the score of model after training
model_1.score(X_train, y_train)
```

```python
# here we are checking the score of model after testing.
model_1.score(X_test, y_test)
# iska score 0.7 se kam hai iska matlab hai hamara model bad hai.
```
