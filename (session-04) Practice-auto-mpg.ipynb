```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets-main/auto-mpg.csv')
df.head()
```

```python
import plotly.express as px
fig=px.pie(df,names='cylinders')
fig.show()
```

```python
import plotly.graph_objects as go
explode_effect=[0.1, 0.1, 0]
custom_colors=['red','green']
fig=go.Figure(go.Pie(
    labels=df['cylinders'],
    pull=explode_effect,
    marker=dict(colors=custom_colors)
))

# set the title
fig.update_layout(title='Titanic passenger class distribution.')

# show the pie chart
fig.show()
```

```python
import plotly.express as px
fig=px.box(df,y='cylinders')
fig.show()
```

```python
import plotly.express as px
fig=px.histogram(df,x='cylinders')
fig.show()
```
