```python
# ismein pie chart ke ek piece ko alag krke dikhaya jata hai. jo piece alag hota hai usi ka color change hoga bas.
# ultimately ismein hum pie chart ke color ko change kr skte hain.
import plotly.graph_objects as go
explode_effect=[0.1, 0.1, 0.1]
# explode ka matlab piechart ke pieces ko alag alag karna.
# aur jo explode variable mein values di hain unhe change krke ham pieces ke bich ke gap ko kam-jada bhi kar sakte hain.
# most important explode ki value sirf or sirf 0 aur 1 ke bich hi honi chahiye (either int or float but b/w 0 and 1).
custom_colors=['yellow']
# baki ke colors same hi rahenge jo pie chart mein bydefault aaye hain.
# 'custom_colors' variable mein agar 1 value bhi dete hain (matlab 1 color dete hain) to bhi koi dikkat nahi hai.
fig=go.Figure(go.Pie(
    labels=df['Gender'],
# labels variable mein apni table ka jo column name denge vo hi pie chart ke rup mein show hoga.
# jese isse phle 'Products' ko labels mein diya tha to products ko pie chart ke rup mein show kr diya.
    pull=explode_effect,
    marker=dict(colors=custom_colors)))

#set the title
fig.update_layout(title='Titanic passenger class distribution.')

# show the pie chart
fig.show()
```

```python
import plotly.graph_objects as go
explode_effect=[0.1, 0.1, 0]
custom_colors=['red','green']
fig=go.Figure(go.Pie(
    labels=df['Product'],
    pull=explode_effect,
    marker=dict(colors=custom_colors)
))

# set the title
fig.update_layout(title='Titanic passenger class distribution.')

# show the pie chart
fig.show()
```
